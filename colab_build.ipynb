```
# ── Шаг 1: Принять лицензии Android SDK (фикс главной проблемы)
import subprocess, os

sdk_root = '/usr/lib/android-sdk'
licenses_dir = f'{sdk_root}/licenses'
os.makedirs(licenses_dir, exist_ok=True)

# Записываем все нужные лицензии вручную
licenses = {
    'android-sdk-license': [
        '8933bad161af4178b1185d1a37fbf41ea5269c55',
        'd56f5187479451eabf01fb78af6dfcb131a6481e',
        '24333f8a63b6825ea9c5514f83c2829b004d1fee',
    ],
    'android-sdk-preview-license': [
        '84831b9409646a918e30573bab4c9c91346d8abd',
    ],
    'intel-android-extra-license': [
        'd975f751698a77b662f1254ddbeed3901e976f5a',
    ],
}

for fname, hashes in licenses.items():
    with open(f'{licenses_dir}/{fname}', 'w') as f:
        f.write('\n'.join(hashes) + '\n')

print('✅ Лицензии приняты')
```

```
# ── Шаг 2: Установить Java 17 (нужна для Gradle)
!apt-get install -y openjdk-17-jdk-headless > /dev/null 2>&1
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
os.environ['PATH'] = os.environ['JAVA_HOME'] + '/bin:' + os.environ['PATH']
!java -version
```

```
# ── Шаг 3: Скачать и настроить Flutter
import os

FLUTTER_VERSION = '3.22.2'
FLUTTER_DIR = '/opt/flutter'

if not os.path.exists(FLUTTER_DIR):
    print('Скачиваем Flutter...')
    !wget -q https://storage.googleapis.com/flutter_infra_release/releases/stable/linux/flutter_linux_{FLUTTER_VERSION}-stable.tar.xz -O /tmp/flutter.tar.xz
    !tar -xf /tmp/flutter.tar.xz -C /opt/
    print('✅ Flutter распакован')
else:
    print('✅ Flutter уже установлен')

os.environ['PATH'] = f'{FLUTTER_DIR}/bin:' + os.environ['PATH']

# Fix git ownership
!git config --global --add safe.directory {FLUTTER_DIR}

!flutter --version
```

```
# ── Шаг 4: Клонировать репозиторий
import os

# ← ВСТАВЬТЕ СЮДА ССЫЛКУ НА ВАШ GitHub репозиторий
REPO_URL = 'https://github.com/YOUR_USERNAME/dating_app.git'
APP_DIR = '/content/dating_app'

if os.path.exists(APP_DIR):
    import shutil
    shutil.rmtree(APP_DIR)

!git clone {REPO_URL} {APP_DIR}
os.chdir(APP_DIR)
print(f'✅ Репозиторий клонирован: {APP_DIR}')
```

```
# ── Шаг 5: Зависимости
os.environ['PATH'] = f'/opt/flutter/bin:' + os.environ['PATH']
!flutter pub get
```

```
# ── Шаг 6: Сборка APK
print('Собираем APK (5-10 минут)...')
!flutter build apk --release 2>&1

apk_path = 'build/app/outputs/flutter-apk/app-release.apk'
if os.path.exists(apk_path):
    size_mb = os.path.getsize(apk_path) / 1024 / 1024
    print(f'\n✅ APK собран! Размер: {size_mb:.1f} MB')
else:
    print('❌ APK не найден — проверьте ошибки выше')
```

```
# ── Шаг 7: Скачать APK
from google.colab import files
apk_path = 'build/app/outputs/flutter-apk/app-release.apk'
files.download(apk_path)
print('✅ Скачивание начато!')
```
